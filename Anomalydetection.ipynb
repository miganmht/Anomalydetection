```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Assuming your preprocessed dataset is stored in a variable called 'sampledataset'
# Assuming your corresponding labels are stored in a variable called 'sampledataset_labels'

# Create an instance of SelectKBest with the desired number of features to select
k = 50  # Specify the desired number of features to select
selector = SelectKBest(score_func=f_classif, k=k)

# Apply feature selection
selected_features = selector.fit_transform(sampledataset, sampledataset_labels)

# Get the indices of the selected features
selected_feature_indices = selector.get_support(indices=True)

# Create a new dataset with the selected features
selected_dataset = sampledataset.iloc[:, selected_feature_indices]
selected_dataset
```

```python
scaler = MinMaxScaler()

normalized_data = scaler.fit_transform(selected_dataset)
normalized_data
```

# Start training


# Train with standardztion

```python
from sklearn.model_selection import train_test_split
X_train, X_test = train_test_split(standardized_data, test_size=0.2,random_state=42)
y_train, y_test = train_test_split(sampledataset_labels, test_size=0.2, random_state=42)
```

## isoltion forest

```python
from sklearn.ensemble import IsolationForest
clf = IsolationForest(random_state=0)
clf.fit(X_train)
```

```python
# Assuming X_test is your unlabeled dataset
predictions = clf.predict(X_test)
predictions[predictions == -1] = 0
```

```python
# Assuming you have imported the necessary libraries and have the variables y_test and predictions
# Convert y_test and predictions to numpy arrays if they are not already

y_test = np.array(y_test)
predictions = np.array(predictions)

# Calculate the percentage of correct predictions
accuracy = (predictions == y_test).mean() * 100

# Print the accuracy percentage
print(f"Accuracy: {accuracy:.2f}%")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fit the model and get the predicted labels
pred = clf.predict(X_train)

# Create a scatter plot
plt.figure(figsize=(10, 7))

# Plot the points, color them based on their predicted labels
plt.scatter(X_train[:, 0], X_train[:, 1], c=pred)

# Add a color bar
plt.colorbar()

# Show the plot
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

# Assuming X_test is your unlabeled dataset
predictions = clf.predict(X_test)

# Convert predictions to 0 and 1
predictions[predictions == 1] = 0
predictions[predictions == -1] = 1

# Calculate confusion matrix
tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()

# Calculate precision, recall, and F1-score
precision = precision_score(y_test, predictions)*100
recall = recall_score(y_test, predictions)*100
f1 = f1_score(y_test, predictions)*100

# Print metrics
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")

# For anomaly scores, you can use decision_function
anomaly_scores = clf.decision_function(X_test)

# Print anomaly scores
print("Anomaly Scores:")
print(anomaly_scores)
```

```python
from sklearn.metrics import roc_curve, auc

# Assuming you have the true labels of your data stored in 'true_labels'
# Assuming you have the anomaly scores predicted by the model stored in 'anomaly_scores'
# Compute the false positive rate (FPR), true positive rate (TPR), and thresholds

fpr, tpr, thresholds = roc_curve(y_test,anomaly_scores )

# Calculate the Area Under the Curve (AUC)
auc_score = auc(fpr, tpr)

# Plot the ROC curve
plt.plot(fpr, tpr, label='ROC Curve (AUC = %0.2f)' % auc_score)
plt.plot([0, 1], [0, 1], 'k--')  # Plotting the diagonal line for reference
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

## Kmeans

```python
from sklearn.cluster import KMeans
# Fit KMeans on the data
kmeans = KMeans(n_clusters=2)
kmeans.fit(X_train)
```

```python
# Assuming X_test is your unlabeled dataset
predictions = kmeans.predict(X_test)
predictions[predictions == -1] = 0
```

```python
# Assuming you have imported the necessary libraries and have the variables y_test and predictions
# Convert y_test and predictions to numpy arrays if they are not already

y_test = np.array(y_test)
predictions = np.array(predictions)

# Calculate the percentage of correct predictions
accuracy = (predictions == y_test).mean() * 100

# Print the accuracy percentage
print(f"Accuracy: {accuracy:.2f}%")
```

```python
# Calculate distances between each point and its assigned centroid
distances = np.sqrt(((X_train - kmeans.cluster_centers_[kmeans.labels_]) ** 2).sum(axis=1))

# Define a threshold. This might be, for example, the 95th percentile of the distances
threshold = np.percentile(distances, 95)

# Anything with distance greater than the threshold is an outlier
outliers = X_train[distances > threshold]

print("Detected outliers:")
print(outliers)
```
